```python
def separate_Apsim_name_details(nameStr):
    '''
    Separate the APSIM name into separate parameters
    This takes a string and separates it into longitude, latidude, variety and sowdate
    
    EG:  nameStr = "11520_-2935_agt_scythe_1-apr"
         returns:  long:  115.20
                   lat:  -29.35
                   variety:  agt_scythe
                   sowDate:  1-apr
    '''
    
    long = 0
    lat = 0
    variety = ""
    sowDate = ""
    
    if nameStr.length > 0:
        #get the first _underscore - this separates the long and lat
        index1 = nameStr.find("_")
        #print(index1)
        long = nameStr[0:index1]
        long = long[0:len(long)-2] + "." + long[-2:]
        #print('long: ', long)
        index2 = nameStr.find("_", index1+1, len(nameStr))
        #print(index2)
        lat = nameStr[index1+1:index2]
        lat = lat[0:len(lat)-2] + "." + lat[-2:]
        #print('lat: ', lat)
        indexE = nameStr.rfind("_")
        #print(indexE)
        variety = nameStr[index2+1: indexE]
        #print('variety: ', variety)
        sowDate = nameStr[indexE+1:]
        #print('sowDate: ', sowDate)        

    return long, lat, variety, sowDate
```
