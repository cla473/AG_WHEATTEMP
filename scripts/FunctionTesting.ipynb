```python
import pandas as pd
import numpy as np
```

```python
apsim_outfiledir = "/OSM/CBR/AG_WHEATTEMP/work"
```

```python
filename = apsim_outfiledir + "/sim57_zadok.csv"
simData = pd.read_csv(filename)
print(simData.shape)
simData
```

```python
simData['avgTemp'] = simData['avgt'].where((simData['date'] >= simData['sowdate']) & (simData['ZadokStage'] > 0) & (simData['ZadokStage'] <= 70), 0)
simData['cumAvgTemp'] = simData.groupby(by=['SimID','sowdate'])['avgTemp'].cumsum()
simData
```

```python
filename = apsim_outfiledir + "/sim57_zadok2.csv"
simData.to_csv(filename, encoding='utf-8', index=False)
```
